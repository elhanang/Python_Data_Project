# Exploratory Data Analysis (EDA)

## Roles to Explore

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Filter Down for US Data Analyst Roles

### What Country has the Highest Number of Job Postings ?

```python
df_plot = df['job_country'].value_counts().to_frame().head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:g_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df['job_title_short'].value_counts().to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='dark:g_r', legend=False)
sns.despine()
plt.title('Distribution of Job Postings by Role')
plt.xlabel('Number of Job Postings')
plt.ylabel('Job Title')
plt.show()
```

## Countries to Explore - **United States** Specifically

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]

df_DA_US
```

## Top 10 Locations in the **US**  (City & States)

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:g_r', legend=False)
sns.despine()
plt.title('Top 10 Job Locations for Data Analysts in the US')
plt.xlabel('Number of Job Postings')
plt.ylabel('Location')
plt.show()
```

## Work-From-Home Availability, Degree Requirements, and Health Insurance Coverage

```python
# Plotting a pie chart for multiple columns

fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), autopct='%1.1f%%', startangle=90, labels=['False', 'True'])
    ax[i].set_title(title)

plt.show()    
```

## What Companies Have the Most Job Postings

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:g_r', legend=False)
sns.despine()
plt.title('Top 10 Companies for Data Analysts in the US')
plt.xlabel('Number of Job Postings')
plt.ylabel('Location')
plt.show()
```
