# 6. Exercise - Analyzing Data Science Job Demand

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) # cleaning up the job_posted_date column. Converting to datetime
```

```python
df_US = df[df['job_country'] == 'United States'].copy()  # Filtering for US jobs only 
```

```python
df_US
```

```python
# Create a new column of the month the job was posted and extract month name
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

df_US
```

```python
# Pivot our data in order to get it into a format suitable for plotting. We're aggregating by count of job titles over time (months).
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short',aggfunc='size')

df_US_pivot
```

```python
# Sort our job posted month in chronological order. Resetting index to turn the month into a column for sorting.
df_US_pivot.reset_index(inplace=True)

# Create a new column to help with sorting the months chronologically
df_US_pivot['month_no'] =pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

df_US_pivot['month_no']
```

```python
df_US_pivot
```

```python
# Sort values by month number
df_US_pivot.sort_values('month_no', inplace=True)

# Set the index
df_US_pivot.set_index('job_posted_month', inplace=True)

# Drop the month_no column as it's no longer needed
df_US_pivot.drop('month_no', axis=1, inplace=True)
```

```python
df_US_pivot 
```

```python
# Plotting the data
df_US_pivot.plot(kind='line')
```

```python
# Cleanup to only have the top three job titles that heav the most postings.
top_3 = df_US['job_title_short'].value_counts().head(3)

top_3 = top_3.index.tolist()

top_3
```

```python
# Plotting the data
df_US_pivot[top_3].plot(kind='line')
plt.title('Top 3 Job Titles Posted Over Months in the US')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.legend(title='Job Titles')
plt.show()
```

# 7. # Merge DataFrames

Basically combining DataFrames when the data is incomplete with new data

pd.merge()

```python
# Import additional data for software development jobs to compare the data science jobs to. 
pd.read_csv("https://lukeb.co/software_csv")
```

```python
# Set the index to job_posted_month so it tallies with the previous data and setting it to a variable.
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')

df_US_software_pivot
```

```python
# Merge the two dataframes to compare data science and software development job trends and set it to a new variable.
df_Us_Merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
# Create a list of the top 5 jobs based on count of jobs in the year.
top_5 = (df_Us_Merged
         .sum() # Sum the counts for each job title across all months
         .sort_values(ascending=False) # Sort the job titles based on total counts
         .head(5) # Get the top 5 job titles based on total counts
         .index # Get the index (job titles) of the top 5
         .to_list() # Convert the index to a list
         )


top_5
```

```python
# Plot the top 5 jobs based on count of jobs in the year and see how they trend over time.
df_Us_Merged[top_5].plot(kind='line')
plt.title('Top 5 Job Titles Posted Over Months in the US in 2023')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.ylim(0, 20000)
plt.legend(title='Job Titles')
plt.show()
```
