# Pie Plots

Plotting Pie Charts and when best to use them

```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) # cleaning up the job_posted_date column. Converting to datetime
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])  # Filling NaN values in the job_skills column with empty lists represented as strings.
```

```python
df.head()
```

```python
# Creating a Pie Chart for Work From Home Options

df['job_work_from_home'].value_counts().plot(kind='pie')
plt.title('Work From Home Options in Job Postings')
plt.ylabel('')  # Hides the y-axis label for better aesthetics
plt.show()
```

```python
# Creating a Pie Chart for Job Titles

df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job Titles in Job Postings')
plt.ylabel('')  # Hides the y-axis label for better aesthetics
plt.show()
```

```python
# Creating a Horizontal Bar Chart for Job Titles

df['job_title_short'].value_counts().plot(kind='barh')
plt.title('Job Titles in Job Postings')
plt.gca().invert_yaxis()  # Inverts the y-axis to have the highest values on top
plt.ylabel('')  # Hides the y-axis label for better aesthetics
plt.show()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%')
plt.title('Work From Home Options in Job Postings')
plt.ylabel('')  # Hides the y-axis label for better aesthetics
plt.show() 
```

```python
# Plotting a pie chart for multiple columns

fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), autopct='%1.1f%%', startangle=90, labels=['False', 'True'])
    ax[i].set_title(title)
```
