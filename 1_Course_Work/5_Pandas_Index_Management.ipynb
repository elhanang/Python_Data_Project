```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) # cleaning up the job_posted_date column. Converting to datetime
```

```python
df.sample(5)  # Displaying a random sample of 5 rows from the DataFrame to verify data loading and cleanup
```

```python
# Inspect the index, which is just numbers. It's a RangeIndex going frpm 0 to the last row.
df.index
```

```python
# Inspecting the name of the index. It has no name currently.
df.index.name = 'job_index' # Attributes is basically just a variable that is associated with an object. Here, name is an attribute of the index object; or just a variable holder. A name, in this case, 'job_index' can be assigned to the index.
```

# Methods to Cover

- df.reset_index()
- df.set_index()
- df.sort_index()

```python
# df.reset_index() Filter for jobs particular to a particular place. It basically resets the index to default integer index.
df_usa = df[df['job_location'] == 'United States']

df_usa.head()
```

```python
df_usa.reset_index(inplace=True) # Resetting the index inplace so that the changes are made to the existing DataFrame rather than creating a new one. We've kept the job_index, but we now have a new default integer index.

df_usa
```

```python
# df.set_index('job_index') Setting the job_index column as the index for the df_usa DataFrame. You lose data sometimes when you use set_index, so it's better to do it inplace.
df_usa.set_index('job_index', inplace=True)
```

```python
df_usa
```

```python
# df.sort_index() Returns a new DataFrame sorted by a label. It's sorting the index, so you don't need to specify the column name.

median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
```

```python
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)

median_pivot
```
