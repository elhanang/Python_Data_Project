```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) # cleaning up the job_posted_date column. Converting to datetime
```

```python
# .iloc to view specific rows and columns. [] index-based selection
df.iloc[0:10, 5]  # Displaying first 10 rows and first 5 columns to verify changes    
```

```python
# .loc method to access a group of rows and columns by labels or a boolean array
df.loc[0:10, 'salary_rate':'salary_hour_avg'] # Displaying first 10 rows and columns from salary_rate to salary_hour_avg
```

```python
# Using .loc with dropna() to remove rows with NaN values in the specified range
df.loc[:, 'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')  # Removing rows with NaN values in salary_rate column
```
