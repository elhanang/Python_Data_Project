```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) # cleaning up the job_posted_date column. Converting to datetime
```

```python
# .sample() to get an overview of the data
df.sample(10) # Displaying 10 random rows from the dataframe to get an overview of the data.
```

```python
# .sample() to get an overview of the data with reproducibility
df.sample(10, random_state=42) # Displaying 10 random rows from the dataframe to get an overview of the data. Setting random_state for reproducibility.
```

# .copy()

```python
# .copy() Make a copy of object's indices and data. It essentially creates a copy of the dataframe.

# Make a new dataframe and load the data into it.
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy() # Creating a copy of the original dataframe to preserve the original data.
```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
# Calculate the median salary and set equal to a variable, median_salary
median_salary = df['salary_year_avg'].median() # Calculating the median salary from the salary_year_avg column.

median_salary
```

```python
# Fill in the null values in that column with the median_salary variable

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary) # Filling in the null values in the salary_year_avg column with the median_salary variable.

df_altered.loc[:5, 'salary_year_avg']
```

```python
# The original gets updated just as the original dateframe did.
df_altered.loc[:5, 'salary_year_avg']
```

```python
# Return the identity of the original dataframe to show it has not changed
id(df_original)
```
