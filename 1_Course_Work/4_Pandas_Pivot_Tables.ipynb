```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date']) # cleaning up the job_posted_date column. Converting to datetime
```

# .pivot_table()

```python
# Count of Job Titles
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
df.groupby('job_title_short').size()
```

```python
# GroupingBy 'Job Title Short' column and Aggregating it to get the median yearly salary
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

## Median Salary Analysis by Country

```python
# Pivot Table with Multiple Indexes and Columns
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median') 
```

```python
# Aggregating to find Top 6 Countries by Job Postings
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist', 'Machine Learning Engineer']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar', figsize=(12, 6))
plt.title('Median Yearly Salary by Job Title and Country')
```
